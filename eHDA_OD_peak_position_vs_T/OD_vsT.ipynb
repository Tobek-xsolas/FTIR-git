```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
A1=np.load('TE108A1.npy')
```

```python
A2=np.load('TE108A2.npy')
```

```python
B1=np.load('TE108B1.npy')
```

```python
B2=np.load('TE108B2.npy')
```

```python
include=[A1,A2,B1,B2]

fig=plt.figure(figsize=(20,10))
for i in include:
    plt.plot(i[0],i[1])
    plt.scatter(i[0],i[1])
    
plt.title('Uncoupled OD peak position in eHDA (TE108) vs Temperature')
plt.xlabel('Temperature [K]')
plt.ylabel('Wavenumber [per cm]')

plt.legend(['A1', 'A2', 'B1', 'B2'])
```
