```python
#init

import numpy as np   #numpy lets us do nice vector manipulations
import glob          #access the file system (for searching for files)
import matplotlib.pyplot as plt          #for plotting
from scipy.signal import savgol_filter   #nice filter
import mpld3                             #toolbar for zooming in plots
mpld3.enable_notebook()
```

```python
#import data

start=3300
end=4000

#Load data (all .ASC files in directory)
files=sorted(glob.glob("*.ASC"))

#Get x axis wavenumbers from a file
x = np.loadtxt(files[1], skiprows=25, unpack=True)[0][start:end]

n=len(files)            #number of files
m=end-start             #data length of each spectrum
data=np.ndarray((n,m))  #create nxm array to hold all data

for i in range(n):
    data[i] = np.loadtxt(files[i], skiprows=25, unpack=True)[1][start:end]
```

```python
#Smooth data

nsdata=np.ndarray((n,m))   #another array to hold processed data

for i in range(n):
    nsdata[i]=savgol_filter(data[i],37,3)
```

```python
#Find peaks (in smoothed data)

peaks=[]

for i in nsdata:
    peaks.append(np.argmax(i[100:300])+100)    #find peak
```

```python
#Normalize

for i in range(n):
    nsdata[i]-=min(data[i])   #subtract minimum
    #sum=np.sum(data[i])
    #data[i]/=sum   #divide by sum (area normalization)
    nsdata[i]/=nsdata[i][peaks[i]]
```

```python
fig=plt.figure(figsize=(15,10))   #creates a figure and sets size

include=[0,1,2,3,4,7,8,9]       #These spectra will be plotted

legend=[str(files[i])+' (' + str(x[peaks[i]])+' cm^-1)' for i in include]   #Use filenames for legend   
#legend=['80 K', '10000 K', 'etc']   #custom legend

c=0
for i in include:
    #plt.annotate(x[peaks[i]], xy=(x[peaks[i]],nsdata[i][peaks[i]]), xytext=(10,10), textcoords='offset points', arrowprops=dict(arrowstyle= '->'))
    plt.plot(x,nsdata[i], linestyle='-')
    #plt.text(x[peaks[i]], nsdata[i][peaks[i]]+(1+c)/30, x[peaks[i]])
    c+=1

plt.legend(legend)
plt.title('Uncoupled OD stretch, eHDA 1% HOD in H2O')
plt.ylabel('Absorbance [arb. u]')
plt.xlabel('Wavenumber [per cm]')
```
